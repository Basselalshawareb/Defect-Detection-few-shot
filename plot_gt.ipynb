```python
import mmcv, cv2, json, matplotlib.pyplot as plt
```

```python
file = json.load(open("/home/huemorgen/DefectDetection/datasets/NEU_DET/base/annotations/test.json"))
```

```python
ann = file['annotations'][800]
```

```python
id = ann['image_id']
```

```python
id
```

```python
chosen_img = None
for img in file['images']:
    if img['id']==id:
        chosen_img = img
        break
```

```python
chosen_img
```

```python
img_to_show = mmcv.imread(f"datasets/NEU_DET/images/{chosen_img['file_name']}")
```

```python
bbox = ann['bbox']
pt1 = (bbox[0], bbox[1])
pt2 = (bbox[0] +bbox[2], bbox[1]+bbox[3])
img_with_gt = cv2.rectangle(img_to_show,pt1, pt2, [0,255,0],3)
```
